This script builds a correlation matrix using the combined dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read in data
data = pd.read_csv(r'/Users/gabegreenberg/Documents/UW/ds_785_capstone/data/historical_data.csv')
print(data)
```

```python
# referenced https://www.datacamp.com/tutorial/seaborn-heatmaps
data = data.drop(columns=['time'])

cm = data.corr()
print("Corelation with isoprene:", cm['isoprene'])

plt.figure(figsize=(15, 12))
sns.heatmap(cm, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
